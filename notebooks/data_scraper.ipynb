```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# pip install bs4
```

```python
# NBA season we will be analyzing
year = 2019
# URL page we will scraping (see image above)
url_avg = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html".format(year)
url_tot = "https://www.basketball-reference.com/leagues/NBA_{}_totals.html".format(year)
# this is the HTML from the given URL
html = urlopen(url_avg)
soup = BeautifulSoup(html)


```

```python
soup.findAll('tr', limit=2)
headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
headers = headers[1:]
headers
```

```python
rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
```

```python
stats = pd.DataFrame(player_stats, columns = headers)
stats.head(10)
```

```python
stats.info
```

```python
url = "https://basketball.fantasysports.yahoo.com/nba/8075/players"
```

```python
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
soup
```

```python
soup.findAll('tr', limit=2)
headers = [th.getText() for th in soup.findAll('tr', limit=2)[1].findAll('th')]
headers
```

```python
rows = soup.findAll('tr')[2:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
```

```python
player_stats[0]
```

```python
stats = pd.DataFrame(player_stats, columns = headers)
stats.head(10)
```

```python
stats.columns
```

```python
column_list = ['Players', 'Roster Status', 'GP*', 'Pre-Season',
       'Current', '% Rostered', 'MPG', 'FGM/A*', 'FG%', 'FTM/A*', 'FT%',
       '3PTM', 'PTS', 'REB', 'AST', 'ST', 'BLK', 'TO']
```

```python
stats=stats[column_list]
```

```python
stats_25 = stats.head(25)
```
